## Load data

```python
base_dir = '/home/hadi/Documents/MTMST'
load_dir = pjoin(base_dir, 'MTLFP', 'xtracted_python')
file = pjoin(load_dir, 'MTLFP_tres25.h5')
file = h5py.File(file, 'r')
```

## Loop

```python
# HIDE CODE


thres_center = 10
thres_opticflow = 1e-1
df = []
for name in file:
    print(name)
    group = file[name]
    nx = np.array(group['nx']).item()
    ny = np.array(group['ny']).item()
    sres = np.unique(group['spatres']).item()
    assert nx == ny 

    hf = np.array(group['hyperflow'])
    hf = HyperFlow(
        opticflow=hf[:, 2:],
        center=hf[:, :2],
        size=nx,
        sres=sres,
        radius=7,
    )
    cutoff_center = []
    cutoff_opticflow = []
    f, px = hf.show_psd('center', False)
    for x in px:
        max_i = np.max(np.where(x > thres_center)[0])
        cutoff_center.append(np.round(f[max_i], 2))
    f, px = hf.show_psd('opticflow', False, cutoff=5)
    for x in px:
        max_i = np.max(np.where(x > thres_opticflow)[0])
        cutoff_opticflow.append(np.round(f[max_i], 2))

    df.append({
        'name': [name] * 2,
        'label': ['center'] * 2,
        'idx': range(2),
        'sd': hf.center.std(0),
        'mu': hf.center.mean(0),
        'cutoff': cutoff_center,
    })
    df.append({
        'name': [name] * 6,
        'label': ['opticflow'] * 6,
        'idx': range(6),
        'sd': hf.opticflow.std(0),
        'mu': hf.opticflow.mean(0),
        'cutoff': cutoff_opticflow,
    })
    msg = ', '.join([
        name,
        f"cutoffs (center): {cutoff_center}",
        f"cutoffs (opticflow): {cutoff_opticflow}",
    ])
    print(msg)
    print('_' * 70)
    print('\n')

df = pd.DataFrame(merge_dicts(df))
print('Done.')
file.close()
```

```python
# HIDE CODE


fig, axes = create_figure(3, 6, (2.4 * 6, 6.8), 'row', 'row')
for i in range(6):
    _df = df.loc[
        (df['idx'] == i) &
        (df['label'] == 'opticflow')
    ]
    ax = axes[0, i]
    sns.histplot(
        data=_df,
        x='sd',
        color=f"C{i}",
        bins=np.linspace(0, 8, 17),
        ax=ax,
    )
    ax.locator_params(axis='x', nbins=10)

    ax = axes[1, i]
    sns.histplot(
        data=_df,
        x='mu',
        color=f"C{i}",
        bins=np.linspace(-0.2, 0.2, 15),
        ax=ax,
    )
    
    ax = axes[2, i]
    sns.histplot(
        data=_df,
        x='cutoff',
        color=f"C{i}",
        bins=np.linspace(0, 7, 15),
        ax=ax,
    )
    ax.locator_params(axis='x', nbins=8)
    
for ax in axes.flat:
    ax.set_xlabel('')
    ax.set_ylabel('')
axes[0, 0].set_ylabel(r'$\sigma$' + '\n(count)')
axes[1, 0].set_ylabel(r'$\mu$' + '\n(count)')
axes[2, 0].set_ylabel('cutoff' + '\n(count)')
fig.suptitle('opticflow', fontsize=15)
plt.show()

print('_' * 70)

fig, axes = create_figure(3, 2, (2.3 * 2, 6.8), 'row', 'row')
for i in range(2):
    _df = df.loc[
        (df['idx'] == i) &
        (df['label'] == 'center')
    ]
    ax = axes[0, i]
    sns.histplot(
        data=_df,
        x='sd',
        color=f"C{i}",
        bins=np.linspace(0, 5, 11),
        ax=ax,
    )
    ax.locator_params(axis='x', nbins=6)

    ax = axes[1, i]
    sns.histplot(
        data=_df,
        x='mu',
        color=f"C{i}",
        bins=np.linspace(-1.2, 1.2, 20),
        ax=ax,
    )
    
    ax = axes[2, i]
    sns.histplot(
        data=_df,
        x='cutoff',
        color=f"C{i}",
        bins=np.linspace(0, 0.2, 11),
        ax=ax,
    )
    ax.locator_params(axis='x', nbins=6)
    
for ax in axes.flat:
    ax.set_xlabel('')
    ax.set_ylabel('')
axes[0, 0].set_ylabel(r'$\sigma$' + '\n(count)')
axes[1, 0].set_ylabel(r'$\mu$' + '\n(count)')
axes[2, 0].set_ylabel('cutoff' + '\n(count)')
fig.suptitle('center', fontsize=15)
plt.show()
```
