## Setup

````python
from model.dataset import setup_repeat_data, setup_supervised_data
from analysis.linear import LinearModel

cfg = ConfigVAE(save=False)
file = h5py.File(cfg.h_file, 'r')
````

````python
hf_kws = {
    'size': 19,
    'sres': 1,
    'radius': 6,
}
model_kws = {
    'category': 'PoissonRegressor',
    'alphas': [1, 10, 50, 100],
    'seed': cfg.seed,
    'verbose': False,
    'n_folds': 5,
}
lags = 16
````

## Loop

````python
%%time

df = []
for expt, grp in file.items():
    hf = HyperFlow(
        params=np.array(grp['hyperflow'])[:, 2:],
        center=np.array(grp['hyperflow'])[:, :2],
        **hf_kws,
    )
    stim = hf.compute_hyperflow()
    spks = np.array(grp['spks'], dtype=float)
    good = ~np.array(grp['badspks'])
    good = np.where(good)[0]

    src, tgt = setup_supervised_data(lags, good, stim, spks)
    stim_r, spks_r, intvl = setup_repeat_data(grp, lags, hf_kws=hf_kws)

    print('_' * 80)
    print(f"fitting {expt} . . .")
    for i in range(tgt.shape[-1]):
        kws = {
            'x': src,
            'y': tgt[:, i],
            'x_tst': stim_r,
            'y_tst': None if spks_r is None
            else np.nanmean(spks_r[i], 0),
            **model_kws,
        }
        glm = LinearModel(**kws).fit()
        nnll = dict(glm.df['nnll'])
        r = dict(glm.df['r'])
        if glm.r2_tst:
            r2 = {
                k: v if v > 0 else np.nan for
                k, v in glm.r2_tst.items()
            }
        else:
            r2 = {
                a: np.nan for
                a in glm.alphas
            }
        if expt in cfg.useful_cells:
            if i in  cfg.useful_cells[expt]:
                useful = 'YES'
            else:
                useful = 'NO'
        else:
            useful = 'NA'
        df.append({
            'expt': [expt] * len(glm.alphas),
            'channel': [i] * len(glm.alphas),
            'useful': [useful] * len(glm.alphas),
            'alpha': glm.alphas,
            'nnll': nnll.values(),
            'r2': r2.values(),
            'r': r.values(),
        })
        # show fits
        _ = glm.show_pred()
        for a, k in glm.kers.items():
            msg = f"expt = {expt} (i = {i})"
            msg += "  --->  "
            msg += ', '.join([
                f"a = {a:0.1f}",
                f"nnll = {nnll[a]:0.3f}",
                f"r = {r[a]:0.3f}",
                f"r2 = {glm.r2_tst.get(a, np.nan):0.1f} %",
            ])
            print(msg)
            k.show()
        # model selection
        if glm.r2_tst:
            best_a, best_r2 = max(
                glm.r2_tst.items(),
                key=lambda t: t[1],
            )
            best_nnll = nnll.get(best_a)
        else:
            best_r2 = np.nan
            best_a, best_nnll = max(
                nnll.items(),
                key=lambda t: t[1],
            )
        best_r = r.get(best_a)
    print(f"fitting {expt}: Done! (best alpha: {best_a} / r: {best_r:0.4f})")
    print('_' * 80)
    print('\n\n')
````
